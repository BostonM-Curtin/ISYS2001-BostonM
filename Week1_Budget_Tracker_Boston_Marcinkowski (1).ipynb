## Part 3: Building the Budget Tracker

Now for the real thing. Ask AI to help you write a program that:
- Takes the user's monthly income
- Takes up to 3 expenses
- Calculates what's left over
- Warns the user if the budget goes negative

> 💡 **Pro tip:** The more specific your prompt, the less clean-up you'll need to do afterwards. "Write a budget tracker" gets you something generic; describing the exact inputs, outputs, and edge cases gets you something closer to what you actually need.


```python
# Write your detailed prompt here:
"""
My Budget Tracker Prompt:
Write a Python program that:
1. Asks the user for their total monthly income.
2. Asks for the names and costs of 3 different monthly expenses.
3. Calculates the total expenses and the remaining balance.
4. Displays a summary of the income, expenses, and balance.
5. If the balance is less than 0, prints a warning message.
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Input: Monthly Income
income = float(input("Enter your total monthly income: $"))

# Input: 3 Expenses
exp1_name = input("Enter the name of expense 1: ")
exp1_cost = float(input(f"Enter the cost of {exp1_name}: $"))

exp2_name = input("Enter the name of expense 2: ")
exp2_cost = float(input(f"Enter the cost of {exp2_name}: $"))

exp3_name = input("Enter the name of expense 3: ")
exp3_cost = float(input(f"Enter the cost of {exp3_name}: $"))

# Calculations
total_expenses = exp1_cost + exp2_cost + exp3_cost
remaining_balance = income - total_expenses

# Summary Output
print("\n--- Budget Summary ---")
print(f"Monthly Income: ${income:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Remaining Balance: ${remaining_balance:,.2f}")

# Warning
if remaining_balance < 0:
    print("\n⚠️ WARNING: You are over budget! Your expenses exceed your income.")
else:
    print("\n✅ Great job! You are within your budget.")
```

```python
# REFINED VERSION — your working budget tracker
income = float(input("Enter your total monthly income: $"))

# Expenses collection
expenses = []
for i in range(1, 4):
    name = input(f"Enter the name of expense {i}: ")
    cost = float(input(f"Enter the cost of {name}: $"))
    expenses.append((name, cost))

# Calculations
total_expenses = sum(cost for name, cost in expenses)
remaining_balance = income - total_expenses

# Display Results
print("\n--- Final Budget Summary ---")
print(f"Total Income:   ${income:,.2f}")
print(f"Total Expenses: ${total_expenses:,.2f}")
print(f"Remaining:      ${remaining_balance:,.2f}")

if remaining_balance < 0:
    print("\n⚠️ Warning: You have exceeded your budget!")
else:
    print("\n✅ You are currently within your budget limit.")
```

## Part 4: Enhancement Challenge

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 A detailed budget summary with advice


```python
# Document your enhancement choice and the prompt you used
enhancement_choice = "📊 Percentage of income spent"
enhancement_prompt = """
Modify my budget tracker code to:
1. Calculate the percentage of the total income that is spent on total expenses.
2. Print this percentage formatted to one decimal place.
3. Add a specific warning if the user is spending more than 80% of their income.
"""

print(f"🚀 Enhancement: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
income = float(input("Enter your total monthly income: $"))

# Expenses collection
expenses = []
for i in range(1, 4):
    name = input(f"Enter the name of expense {i}: ")
    cost = float(input(f"Enter the cost of {name}: $"))
    expenses.append((name, cost))

# Calculations
total_expenses = sum(cost for name, cost in expenses)
remaining_balance = income - total_expenses

# NEW: Calculate Percentage
if income > 0:
    percent_spent = (total_expenses / income) * 100
else:
    percent_spent = 0

# Display Results
print("\n--- Enhanced Budget Summary ---")
print(f"Total Income:      ${income:,.2f}")
print(f"Total Expenses:    ${total_expenses:,.2f}")
print(f"Remaining:         ${remaining_balance:,.2f}")
print(f"Percentage Spent:  {percent_spent:.1f}%")

if remaining_balance < 0:
    print("\n⚠️ Warning: You have exceeded your budget!")
elif percent_spent > 80:
    print("\n⚠️ Warning: You are spending over 80% of your income. Consider saving more!")
else:
    print("\n✅ You are currently within a healthy budget limit.")
```
